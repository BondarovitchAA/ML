```python
import sklearn
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris.data.shape
```

```python
iris_embeded = TSNE(n_components = 2, perplexity=5, n_iter = 5000).fit_transform(iris.data)
iris_embeded.shape
```

```python
plt.figure(figsize=(12, 12))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)
plt.subplot(321)
plt.title("t-SNE, perplexity = 5", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
iris_embeded = TSNE(n_components = 2, perplexity=10, n_iter = 5000).fit_transform(iris.data)
plt.subplot(322)
plt.title("t-SNE, perplexity = 10", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
iris_embeded = TSNE(n_components = 2, perplexity=20, n_iter = 5000).fit_transform(iris.data)
plt.subplot(323)
plt.title("t-SNE, perplexity = 20", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
iris_embeded = TSNE(n_components = 2, perplexity=30, n_iter = 5000).fit_transform(iris.data)
plt.subplot(324)
plt.title("t-SNE, perplexity = 30", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
iris_embeded = TSNE(n_components = 2, perplexity=40, n_iter = 5000).fit_transform(iris.data)
plt.subplot(325)
plt.title("t-SNE, perplexity = 40", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
iris_embeded = TSNE(n_components = 2, perplexity=50, n_iter = 5000).fit_transform(iris.data)
plt.subplot(326)
plt.title("t-SNE, perplexity = 50", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')
plt.show()
```

Как мы можем заметить, основной проблемой метода t-SNE, является выбор правильного параметра perplexity, что является затруднительным в случае отсутсвия информации о начальном распределении. Наиболее подходящиими параметрами для набора "Ирисы Фишера" оказались perplexity = 5, 20, 30, в этом случае можно четко проследить разделимость классов. При количестве итераций = 5000 алгоритм становится стабильным и уже не подвержен изменениям.

```python
from sklearn.decomposition import PCA
iris_embeded = PCA(n_components = 2).fit_transform(iris.data)
iris_embeded.shape
```

```python
plt.figure(figsize=(12, 12))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)
plt.subplot(421)
plt.title("PCA, whiten = false, svd_solver=auto", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(422)
iris_embeded = PCA(n_components = 2, whiten=True).fit_transform(iris.data)
plt.title("PCA, whiten = true, svd_solver=auto", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(423)
iris_embeded = PCA(n_components = 2, svd_solver='full').fit_transform(iris.data)
plt.title("PCA, whiten = false, svd-solver=full", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(424)
iris_embeded = PCA(n_components = 2, svd_solver='full', whiten=True).fit_transform(iris.data)
plt.title("PCA, whiten = true, svd-solver=full", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(425)
iris_embeded = PCA(n_components = 2, svd_solver='arpack').fit_transform(iris.data)
plt.title("PCA, whiten = false, svd-solver=arpack", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(426)
iris_embeded = PCA(n_components = 2, whiten=True, svd_solver='arpack').fit_transform(iris.data)
plt.title("PCA, whiten = true, svd-solver=arpack", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(427)
iris_embeded = PCA(n_components = 2, svd_solver='randomized').fit_transform(iris.data)
plt.title("PCA, whiten = false, svd-solver=randomized", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.subplot(428)
iris_embeded = PCA(n_components = 2, whiten=True, svd_solver='randomized').fit_transform(iris.data)
plt.title("PCA, whiten = true, svd-solver=randomized", fontsize='small')
plt.scatter(iris_embeded[:, 0], iris_embeded[:, 1], marker='o', c=iris.target,
            s=25, edgecolor='k')

plt.show()
```

Как можно видеть, алгоритм PCA дает четкое разделение всех трех классов. При этом передача различных параметров практически не влияет на результат (по крайней мере на данном примере), что уменьшает вероятность ошибиться в выборе параметров. Также масштаб графиков значительно меньше, по сравнению с t-SNE, что увеличивает визуальную точность и уменьшает дисперсию.

ИТОГО:
Заметим, что график алгоритма t-SNE при параметре perplexity = 5 практически полностью совпадает с графиками PCA алгоритма.
При этом использование различных параметров для этого метода позволяет рассмотреть распределние с разных сторон, что дает бОльшую гибкость по сравнению с PCA.
Для использования на практике следует использовать комбинацию этих алгоритмов. t-SNE позволяет получить более общую картину, но подвержен большим изменениям, из-за чего смотря на графики можно легко запутаться. Алгоритм PCA позволит подтвердить какие-то догадки при использовании t-SNE. Также скорость алгоритма t-SNE значительно медленнее, чем PCA, поэтому для пространств с большой размерностью можно использовать комбинацию алгоритмов.
